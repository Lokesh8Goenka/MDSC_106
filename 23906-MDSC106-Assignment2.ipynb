```python
def pie(dframe,col):                                        # function to print the pir plot for a specific column in a df

    label = df[col].unique()                                # setting the label as the unique values in the column
    l = len(label)                                          # calulating the number of unique values
    counts  = []        
    for i in range(l):                                                  
        counts.append(df[col].value_counts()[i])            # counting the total number of values for the unique values in the column 
                                                            # and then appending the counts to the list 'counts' .

    plt.pie(counts, labels = label, autopct='%1.1f%%')      # plotting the pie plot for the counts of the unique values in the column
    # print(dframe)                                         # eg. gender = male-> 3555, and female-> 3488.
    
```

```python
features = ['gender', 'SeniorCitizen', 'Partner', 'Dependents','PhoneService', 'MultipleLines', 'InternetService','OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport','StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling','PaymentMethod', 'Churn']

plt.figure(figsize=(20,30))                             # setting the plot size 
j = 1                                                   # to set the position of the first plot.

for i in range(len(features)):                          # for loop to run through all the features in the dataframe.
        plt.subplot(6,3,j)                              # setting the position of the plot.  (6 * 3 = 18 plots)
        plt.title("{}".format(features[i]))           
        pie(df,features[i])                             # calling the function pie.
        j += 1                                          # going to the next position

plt.show()        

# The below plot tells us about percentage of all the unique values present in all the columns.
```
