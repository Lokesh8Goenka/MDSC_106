### Diabetes-Logistic Regression

```python
def diagnostic_plots(df,variable,target):
    
    # figure size
    plt.figure(figsize=(20,7))

    #Histogram
    plt.subplot(1,4,1)
    sns.histplot(df[variable], kde=True, color = 'r')
    plt.title('Histogram')

    # Scatter
    plt.subplot(1,4,2)
    plt.scatter(df[variable], df[target], color = 'g')
    plt.title("Scatterplot")

    # Barplot
    plt.subplot(1,4,3)
    sns.barplot(x = target, y = variable, data = df)
    plt.title('Barplot')

    plt.show()
```

```python
diagnostic_plots(df,'bmi','age')
```

### Line plot

* Line graphs are used to track changes over short and long periods of time. 
* When smaller changes exist, line graphs are better to use than bar graphs. 
* Line graphs can also be used to compare changes over the same period of time for more than one group

### Scatter Plot

 * Use a scatter plot to determine whether or not two variables have a relationship or correlation.
 * Plotting a scattergram with your data points can help you to determine whether there is a potential relationship between them

### Box Plot

A box plot consist of 5 things.

Minimum
First Quartile or 25%
Median (Second Quartile) or 50%
Third Quartile or 75%
Maximum

```python
df2 = pd.read_csv('./tips.csv')     # data set from (https://www.geeksforgeeks.org/box-plot-visualization-with-pandas-and-seaborn/)
df2.boxplot(by = 'day', column=['tip'], grid =False)
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = df2)
```

* Bottom black horizontal line of blue box plot is minimum value
* First black horizontal line of rectangle shape of blue box plot is First quartile or 25%
* Second black horizontal line of rectangle shape of blue box plot is Second quartile or 50% or median.
* Third black horizontal line of rectangle shape of blue box plot is third quartile or 75%
* Top black horizontal line of rectangle shape of blue box plot is maximum value.
* Small diamond shape of blue box plot is outlier data or erroneous data.
